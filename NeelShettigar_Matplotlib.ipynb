# Line plots

### Generate a lineplot of y1 and y2 on x axis


```python
# Using x, y1 and y2 
# Generate of y1 and y2 on x axis
x = np.linspace(0, 20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

#Code goes below:
plt.plot(x, y1)
plt.plot(x,y2)
plt.show()
```

## Color and style

### Create same plot above with:
### - y1 color is black, linestyle solid, linewidth of 2, squar marker at a size of 6
### - y2 color is gray, linestyle dashed, linewidth of 2, triangle markers at size of 6



```python
# Using:
x = np.linspace(0, 20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

# Create same plot above with:
# - y1 color is black, linestyle solid, linewidth of 2, squar marker at a size of 6
# - y2 color is gray, linestyle dashed, linewidth of 2, triangle markers at size of 6

# Code goes below
plt.plot(x, y1, c = "black", ls = '-', linewidth = '2', marker = 's', markersize = 6)
plt.plot(x, y2, c = "gray", ls = '--', linewidth = '2', marker = '^', markersize = 6)
plt.show()
```

# Vertical bar charts

### Create a vertical bar chart for the values by years. 
### Its ok to have years split i.e, (2012.0, 2012.5...)


```python
# Using
years = np.arange(2012, 2015)
values = [2, 5, 9]
# Create a vertical bar chart values by years.

# Code goes below:
plt.bar(years, values)
plt.show()
```

## Create a multiserise bar chart with:
* category1_values: Color = Blue, No edgecolor, width = 0.4, align = center and a label of y1
* category2_values: Color = Orange, No edgecolor, width = 0.4, align = center and a label of y2


```python
# Using:
years = np.arange(2012, 2015)
category1_values = [2, 5, 9]
category2_values = [7, 6, 3]

# Create a multiserise bar chart with:
#   category1_values: Color = Blue, No edgecolor, width = 0.4, align = center and a label of y1
#   category2_values: Color = Orange, No edgecolor, width = 0.4, align = center and a label of y2

# Code goes below:
ax = plt.subplot()
ax.bar(years-.2, category1_values, width=0.4, color='blue', align='center', label = "y1")
ax.bar(years+.2,category2_values , width=0.4, color='orange', align='center', label = "y2")
plt.legend()
plt.show()

```

# Horizontal bar charts
### Create a horizontal bar chart for the given values:
* categories = ['A', 'B', 'C', 'D', 'E']
* values = [7, 12, 4, 2, 9]
* Lable each as "Category and letter... i.e. Category A, Category B ...


```python
# Using
categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 12, 4, 2, 9]

# Create a horizontal bar chart for the values
# Lable each as "Category and letter... i.e. Category A, Category B ..

# Code goes below:

fig,ax = plt.subplots()
y_pos = np.arange(len(categories))
ax.barh(y_pos, values, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(['Catagory A', 'Catagory B', 'Catagory C', 'Catagory D', 'Catagory E'])

plt.show()
```

### Create a horizontal bar chart for the given values with:
* Color = Blue, no edgecolor, height of 0.6, center aligned


```python
# Using:
categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 12, 4, 2, 9]

# Create a horizontal bar chart with
# Color = Blue, no edgecolor, height of 0.6, center aligned

# Code goes below:

fig, ax = plt.subplots()
y_pos = np.arange(len(categories))
ax.barh(y_pos, values, height = .6, align='center', color = 'blue')
ax.set_yticks(y_pos)
ax.set_yticklabels(['Catagory A', 'Catagory B', 'Catagory C', 'Catagory D', 'Catagory E'])

plt.show()
```

# Pie charts

## Create a pie chart
* For the given counts = [17, 14]


```python
# Using
counts = [17, 14]

# Create a pie chart

# Code goes below:
plt.pie(counts)
plt.show()
```

### Create a pie chart for counts = [17, 14] using:
* Colors of blue and orange
* Labels of 'Category A' and 'Other categories'
* Set the angle at 90
* Set the percentage display format of 1.1



```python
# Using
counts = [17, 14]

# Create a pie chart for counts = [17, 14] using:
#  Colors of blue and orange
#  Labels of 'Category A' and 'Other categories'
#  Set the angle at 90
#  Set the percentage display format of 1.1

# Code goes below:
plt.pie(counts, labels = ['Category A', 'Other categories'], startangle = 90, colors = ['blue', 'orange'], autopct='%1.1f%%')
plt.show()
```

# Scatter plots

## Create a scatter plot for the given data



```python
# Using
x = range(20)
y = np.arange(50, 70) + (np.random.random(20) * 10.)

# Create a scatter plot for the given data

#Code goes below:
plt.scatter(x, y)
plt.show()
```

## Create a scatter plot with the given data with:
* Color of red
* Marker 

* `c`: Set the color of the markers.

* `s`: Set the size of the markers.

* `marker`: Set the marker style, e.g., circles, triangles, or squares.

* `edgecolor`: Set the color of the lines on the edges of the markers.

```python
x = np.arange(20)
y = np.arange(50, 70) + (np.random.random(20) * 10)

c = 'red'
s = 50
marker = 's'
edgecolor = 'blue'
plt.scatter(x, y, color = c, s = s, marker = marker, edgecolors = edgecolor)
plt.show()
```
