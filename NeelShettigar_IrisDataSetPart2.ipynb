This program will plot and calculate correlation between sepal and petal width by species, and sepal and petal length by species. It will also plot the distribution of sepal lengths by species.

```python
# Import libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read data into DataFrame
df = pd.read_csv('iris.csv')
```

```python
# Correlation between sepal and petal width by species 
sns.FacetGrid(df, hue = 'species').map(plt.scatter,'sepal_width','petal_width').add_legend()
plt.show()
print(df.groupby('species')['sepal_width'].corr(df['petal_width']))
```

# Observations
Setosa has a low correlation of around .28, which is evident in the graph as it appears to be a somewhat horizontal line. The other two species have a correlation between .5 and .7 which indicates a moderate correlation. This is also evident in the graph as both species cluster have a positive somewhat linear relationship between sepal and petal width.

```python
# Correlation between sepal and petal length by species 
sns.FacetGrid(df, hue = 'species').map(plt.scatter,'sepal_length','petal_length').add_legend()
plt.show()
print(df.groupby('species')['sepal_length'].corr(df['petal_length']))
```

# Observations
Setosa has a low correlation of around .26, which is evident in the graph as it appears to be a somewhat horizontal line. The other two species have a correlation between .7 and .9 which indicates a fairly strong correlation. This is also evident in the graph as both species cluster have a positive linear relationship between sepal and petal length.

```python
# Sepal lengths distributed by species 
fig, axs = plt.subplots(3, sharex=True, sharey=True)
axs[1].set(ylabel = 'Frequency')
axs[2].set(xlabel = 'Sepal Length')
axs[0].hist(df[df['species'] == 'setosa']['sepal_length'])
axs[0].set_title('Setosa')
axs[1].hist(df[df['species'] == 'versicolor']['sepal_length'])
axs[1].set_title('Versicolor')
axs[2].hist(df[df['species'] == 'virginica']['sepal_length'])
axs[2].set_title('Virginica')
fig.tight_layout()

# I wanted to group together my graphs so that they would share axes and have consistent scaling,
# so I looked up how to do this and changed the code to include this data set and a histogram
# instead of a regular plot. 
# Reference link: https://matplotlib.org/devdocs/gallery/subplots_axes_and_figures/subplots_demo.html
```

# Observations
Setosa has a sepal length distribution that is unimodal, symmetric, and approximately 'Normal'. Versicolor has a sepal length distribution that is multimodal and approximately 'Horizontal'. Finally, Virginica has a sepal length distribution that is bimodal and 'Skewed left. The means and standard deviations increases by species, with setosa having the smallest mean and standard deviation, and virginica having the greatest mean and standard deviation.
